### <span style="font-family: 'Times New Roman', Times, serif;">Slide 8: What Are Data Types?</span>

<span style="font-family: 'Times New Roman', Times, serif;">

In Python, **everything is an object**. Every value has a type, and you can inspect it using:
- `type()` -- returns the type of an object.
- `isinstance()` -- checks if an object is an instance of a specific type.

</span>

````python
# Everything in Python is an object
print("=== Using type() ===")
print(f"type(42)       = {type(42)}")
print(f"type(3.14)     = {type(3.14)}")
print(f"type('hello')  = {type('hello')}")
print(f"type(True)     = {type(True)}")
print(f"type(None)     = {type(None)}")
print(f"type([1,2,3])  = {type([1,2,3])}")
print(f"type(print)    = {type(print)}")

print("\n=== Using isinstance() ===")
x = 42
print(f"isinstance(42, int):       {isinstance(x, int)}")
print(f"isinstance(42, float):     {isinstance(x, float)}")
print(f"isinstance(42, (int, float)): {isinstance(x, (int, float))}")

# Even functions and classes are objects
print(f"\ntype(type) = {type(type)}")
print(f"id(42) = {id(42)}  (every object has an identity)")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 9: Numeric Types</span>

<span style="font-family: 'Times New Roman', Times, serif;">

Python provides several numeric types:

| Type | Description | Example |
|---|---|---|
| `int` | Arbitrary-precision integers | `42`, `-7`, `10**100` |
| `float` | IEEE 754 double-precision | `3.14`, `1.0e-5` |
| `complex` | Complex numbers | `3+4j` |
| `Decimal` | Fixed-point precision | `Decimal('0.1')` |
| `Fraction` | Exact rational numbers | `Fraction(1, 3)` |

</span>

````python
# Integer examples
a = 42
big_number = 10 ** 100  # Python handles arbitrarily large integers
print(f"a = {a}, type = {type(a)}")
print(f"10^100 = {big_number}")
print(f"Binary: {bin(42)}, Octal: {oct(42)}, Hex: {hex(42)}")

# Float examples
b = 3.14159
scientific = 1.5e-3
print(f"\nb = {b}, type = {type(b)}")
print(f"Scientific notation: {scientific}")
print(f"Float precision issue: 0.1 + 0.2 = {0.1 + 0.2}")

# Complex numbers
c = 3 + 4j
print(f"\nc = {c}, type = {type(c)}")
print(f"Real part: {c.real}, Imaginary part: {c.imag}")
print(f"Magnitude: {abs(c)}")
````

````python
# Decimal and Fraction for precision
from decimal import Decimal
from fractions import Fraction

# Decimal avoids floating-point issues
print("=== Decimal ===")
print(f"Float:   0.1 + 0.2 = {0.1 + 0.2}")
print(f"Decimal: 0.1 + 0.2 = {Decimal('0.1') + Decimal('0.2')}")

# Fraction for exact rational arithmetic
print("\n=== Fraction ===")
f1 = Fraction(1, 3)
f2 = Fraction(1, 6)
print(f"1/3 = {f1}")
print(f"1/6 = {f2}")
print(f"1/3 + 1/6 = {f1 + f2}")
print(f"Fraction from float: {Fraction(0.5)} = 1/2")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 10: Strings</span>

<span style="font-family: 'Times New Roman', Times, serif;">

Strings in Python are **immutable** sequences of characters. Key features:
- Single, double, or triple quotes
- **f-strings** for formatted string literals (Python 3.6+)
- Useful methods: `.join()`, `.split()`, `.replace()`, `.upper()`, `.lower()`, `.strip()`

</span>

````python
# String creation
s1 = 'Single quotes'
s2 = "Double quotes"
s3 = """Triple quotes
can span multiple lines"""

print(s1)
print(s2)
print(s3)

# Strings are immutable
name = "Python"
print(f"\nOriginal: {name}")
# name[0] = 'J'  # This would raise TypeError
new_name = 'J' + name[1:]  # Create a new string instead
print(f"Modified: {new_name}")
````

````python
# f-strings (formatted string literals)
name = "Alice"
age = 25
gpa = 3.856

print(f"Name: {name}")
print(f"Age: {age}")
print(f"GPA: {gpa:.2f}")  # Format to 2 decimal places
print(f"Next year: {age + 1}")
print(f"Name uppercase: {name.upper()}")

# String methods: join, split, replace
print("\n=== String Methods ===")
words = ["Python", "is", "awesome"]
sentence = " ".join(words)
print(f".join(): {sentence}")

csv_data = "apple,banana,cherry"
items = csv_data.split(",")
print(f".split(): {items}")

text = "Hello World"
print(f".replace(): {text.replace('World', 'Python')}")
print(f".upper(): {text.upper()}")
print(f".lower(): {text.lower()}")
print(f".strip(): {'  spaces  '.strip()}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 11: Lists and Tuples</span>

<span style="font-family: 'Times New Roman', Times, serif;">

- **Lists** are **mutable** ordered sequences. Use `[]`.
- **Tuples** are **immutable** ordered sequences. Use `()`.
- **List comprehensions** provide a concise way to create lists.

</span>

````python
# Lists: mutable
print("=== Lists ===")
fruits = ["apple", "banana", "cherry"]
print(f"Original: {fruits}")

fruits.append("date")
print(f"After append: {fruits}")

fruits[1] = "blueberry"  # Mutable: can change elements
print(f"After modify: {fruits}")

fruits.remove("cherry")
print(f"After remove: {fruits}")

# Indexing and slicing
numbers = [10, 20, 30, 40, 50]
print(f"\nSlicing: {numbers[1:4]}")
print(f"Step: {numbers[::2]}")
print(f"Reverse: {numbers[::-1]}")
````

````python
# Tuples: immutable
print("=== Tuples ===")
point = (3, 4)
print(f"Point: {point}, type: {type(point)}")
print(f"x = {point[0]}, y = {point[1]}")

# Tuple unpacking
x, y = point
print(f"Unpacked: x={x}, y={y}")

# Tuples are immutable
# point[0] = 5  # This would raise TypeError

# List comprehensions
print("\n=== List Comprehensions ===")
squares = [x**2 for x in range(1, 11)]
print(f"Squares: {squares}")

evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"Evens: {evens}")

words = ["hello", "world", "python"]
upper_words = [w.upper() for w in words]
print(f"Uppercased: {upper_words}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 12: Dictionaries</span>

<span style="font-family: 'Times New Roman', Times, serif;">

Dictionaries store **key-value pairs**. Since Python 3.7, they maintain **insertion order**.

- Access values with `dict[key]` or `dict.get(key, default)`.
- Specialised dictionaries: `defaultdict`, `Counter` (from `collections`).

</span>

````python
# Basic dictionary operations
student = {
    "name": "Alice",
    "age": 22,
    "major": "Computer Science",
    "gpa": 3.85
}

print(f"Student: {student}")
print(f"Name: {student['name']}")
print(f"GPA: {student.get('gpa', 'N/A')}")
print(f"Phone: {student.get('phone', 'Not provided')}")

# Dictionary methods
print(f"\nKeys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
print(f"Items: {list(student.items())}")

# Adding and updating
student["email"] = "alice@university.edu"
student.update({"year": 3, "gpa": 3.90})
print(f"\nUpdated: {student}")
````

````python
# defaultdict and Counter from collections
from collections import defaultdict, Counter

# defaultdict: provides default values for missing keys
word_count = defaultdict(int)
text = "the cat sat on the mat the cat"
for word in text.split():
    word_count[word] += 1
print(f"defaultdict word count: {dict(word_count)}")

# Counter: built-in counting
counter = Counter(text.split())
print(f"Counter: {counter}")
print(f"Most common 2: {counter.most_common(2)}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 13: Sets and Frozensets</span>

<span style="font-family: 'Times New Roman', Times, serif;">

- **Sets** are unordered collections of **unique elements**. They are mutable.
- **Frozensets** are the immutable version of sets.
- Sets support **set algebra**: union, intersection, difference, symmetric difference.

</span>

````python
# Sets: unique elements
numbers = {1, 2, 3, 3, 4, 4, 5}
print(f"Set (duplicates removed): {numbers}")

# Set operations
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print(f"\nA = {a}")
print(f"B = {b}")
print(f"Union (A | B):        {a | b}")
print(f"Intersection (A & B): {a & b}")
print(f"Difference (A - B):   {a - b}")
print(f"Symmetric Diff (A ^ B): {a ^ b}")

# Set methods
a.add(6)
print(f"\nAfter add(6): {a}")
a.discard(1)
print(f"After discard(1): {a}")
````

````python
# Frozenset: immutable set
fs = frozenset([1, 2, 3, 4])
print(f"Frozenset: {fs}, type: {type(fs)}")
# fs.add(5)  # Would raise AttributeError

# Frozensets can be used as dictionary keys or set elements
set_of_sets = {frozenset([1, 2]), frozenset([3, 4])}
print(f"Set of frozensets: {set_of_sets}")

# Practical use: removing duplicates from a list
names = ["Alice", "Bob", "Alice", "Charlie", "Bob"]
unique_names = list(set(names))
print(f"\nOriginal: {names}")
print(f"Unique: {unique_names}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 15: Type Conversion and Custom Types</span>

<span style="font-family: 'Times New Roman', Times, serif;">

Python provides built-in functions for **type conversion** (also called *type casting*):
- `int()`, `float()`, `str()`, `bool()`, `list()`, `tuple()`, `set()`
- Use `isinstance()` to check types before conversion.

</span>

````python
# Type conversion examples
print("=== Type Conversion ===")

# String to int/float
age_str = "25"
age_int = int(age_str)
print(f"str -> int: '{age_str}' -> {age_int} (type: {type(age_int).__name__})")

price_str = "19.99"
price_float = float(price_str)
print(f"str -> float: '{price_str}' -> {price_float} (type: {type(price_float).__name__})")

# Number to string
num = 42
num_str = str(num)
print(f"int -> str: {num} -> '{num_str}' (type: {type(num_str).__name__})")

# Float to int (truncates)
pi = 3.99
print(f"float -> int: {pi} -> {int(pi)} (truncates, does not round)")

# Sequence conversions
print(f"\nlist -> tuple: {tuple([1, 2, 3])}")
print(f"tuple -> list: {list((1, 2, 3))}")
print(f"list -> set: {set([1, 2, 2, 3, 3])}")
print(f"string -> list: {list('hello')}")
````

````python
# Using isinstance() for type checking
def describe_value(value):
    """Describe a value's type and properties."""
    if isinstance(value, int) and not isinstance(value, bool):
        print(f"{value} is an integer, even: {value % 2 == 0}")
    elif isinstance(value, float):
        print(f"{value} is a float, integer part: {int(value)}")
    elif isinstance(value, str):
        print(f"'{value}' is a string, length: {len(value)}")
    elif isinstance(value, (list, tuple)):
        print(f"{value} is a {type(value).__name__}, length: {len(value)}")
    elif isinstance(value, bool):
        print(f"{value} is a boolean")
    else:
        print(f"{value} is of type {type(value).__name__}")

test_values = [42, 3.14, "hello", [1, 2, 3], (4, 5), True, None]
for val in test_values:
    describe_value(val)
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 18: Variables</span>

<span style="font-family: 'Times New Roman', Times, serif;">

In Python:
- **No type declaration** is needed -- the type is inferred at runtime.
- **Naming rules:**
  - Must start with a letter or underscore (`_`).
  - Can contain letters, digits, and underscores.
  - Case-sensitive (`name` and `Name` are different).
  - Cannot use reserved keywords (`if`, `for`, `class`, etc.).

</span>

````python
# Variable assignment: no type declaration needed
name = "Alice"         # string
age = 25               # int
height = 5.6           # float
is_student = True      # bool
courses = ["Math", "CS"]  # list

print(f"name = {name} (type: {type(name).__name__})")
print(f"age = {age} (type: {type(age).__name__})")
print(f"height = {height} (type: {type(height).__name__})")
print(f"is_student = {is_student} (type: {type(is_student).__name__})")
print(f"courses = {courses} (type: {type(courses).__name__})")

# Multiple assignment
x, y, z = 1, 2, 3
print(f"\nMultiple assignment: x={x}, y={y}, z={z}")

# Same value to multiple variables
a = b = c = 0
print(f"Same value: a={a}, b={b}, c={c}")
````

````python
# Variable naming rules
print("=== Valid Variable Names ===")
my_variable = 1
_private = 2
camelCase = 3
snake_case = 4
name2 = 5

# Case sensitivity
Name = "Alice"
name = "Bob"
NAME = "Charlie"
print(f"Name={Name}, name={name}, NAME={NAME}  (all different!)")

# Python keywords (cannot be used as variable names)
import keyword
print(f"\nPython reserved keywords ({len(keyword.kwlist)}):\n{keyword.kwlist}")
````

| Operator | Operation | Example | Result |
|---|---|---|---|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `10 - 4` | `6` |
| `*` | Multiplication | `6 * 7` | `42` |
| `/` | Division (float) | `15 / 4` | `3.75` |
| `//` | Floor Division | `15 // 4` | `3` |
| `%` | Modulus | `15 % 4` | `3` |
| `**` | Exponentiation | `2 ** 10` | `1024` |

</span>

````python
# Arithmetic Operators
a, b = 15, 4
print(f"a = {a}, b = {b}")
print(f"{'='*35}")
print(f"Addition:       a + b  = {a + b}")
print(f"Subtraction:    a - b  = {a - b}")
print(f"Multiplication: a * b  = {a * b}")
print(f"Division:       a / b  = {a / b}")
print(f"Floor Division: a // b = {a // b}")
print(f"Modulus:        a % b  = {a % b}")
print(f"Exponentiation: a ** b = {a ** b}")

# Practical examples
print(f"\n=== Practical Examples ===")
total_minutes = 137
hours = total_minutes // 60
minutes = total_minutes % 60
print(f"{total_minutes} minutes = {hours} hours and {minutes} minutes")

# Temperature conversion
celsius = 37
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}C = {fahrenheit}F")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 21: Comparison and Logical Operators</span>

<span style="font-family: 'Times New Roman', Times, serif;">

**Comparison Operators** return `True` or `False`:
- `==`, `!=`, `>`, `<`, `>=`, `<=`

**Logical Operators** combine boolean expressions:
- `and`, `or`, `not`

</span>

````python
# Comparison Operators
x, y = 10, 20
print(f"x = {x}, y = {y}")
print(f"{'='*35}")
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x > y:  {x > y}")
print(f"x < y:  {x < y}")
print(f"x >= y: {x >= y}")
print(f"x <= y: {x <= y}")

# Chained comparisons (unique to Python)
age = 25
print(f"\nChained: 18 <= {age} <= 65 -> {18 <= age <= 65}")
print(f"Chained: 1 < 2 < 3 < 4 -> {1 < 2 < 3 < 4}")
````

````python
# Logical Operators
print("=== Logical Operators ===")

age = 25
has_id = True
is_vip = False

# and: both must be True
print(f"age >= 18 and has_id: {age >= 18 and has_id}")
print(f"age >= 18 and is_vip: {age >= 18 and is_vip}")

# or: at least one must be True
print(f"has_id or is_vip: {has_id or is_vip}")
print(f"False or False: {False or False}")

# not: inverts the boolean
print(f"not is_vip: {not is_vip}")
print(f"not True: {not True}")

# Practical example: eligibility check
print("\n=== Eligibility Check ===")
score = 85
attendance = 92
eligible = score >= 80 and attendance >= 90
print(f"Score: {score}, Attendance: {attendance}%")
print(f"Eligible for honors: {eligible}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 22: Assignment and Bitwise Operators</span>

<span style="font-family: 'Times New Roman', Times, serif;">

**Assignment Operators:**
- `=`, `+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`

**Bitwise Operators** operate on individual bits:
- `&` (AND), `|` (OR), `^` (XOR), `~` (NOT), `<<` (left shift), `>>` (right shift)

</span>

````python
# Assignment Operators
print("=== Assignment Operators ===")
x = 10
print(f"x = {x}")

x += 5   # x = x + 5
print(f"x += 5  -> x = {x}")

x -= 3   # x = x - 3
print(f"x -= 3  -> x = {x}")

x *= 2   # x = x * 2
print(f"x *= 2  -> x = {x}")

x /= 4   # x = x / 4
print(f"x /= 4  -> x = {x}")

x = 17
x //= 3  # x = x // 3
print(f"\n17 //= 3  -> x = {x}")

x = 17
x %= 5   # x = x % 5
print(f"17 %= 5   -> x = {x}")

x = 2
x **= 10 # x = x ** 10
print(f"2 **= 10  -> x = {x}")
````

````python
# Bitwise Operators
print("=== Bitwise Operators ===")
a = 0b1100  # 12 in binary
b = 0b1010  # 10 in binary

print(f"a = {a} (binary: {bin(a)})")
print(f"b = {b} (binary: {bin(b)})")
print(f"{'='*40}")

print(f"a & b  (AND):  {a & b:>3}  (binary: {bin(a & b)})")
print(f"a | b  (OR):   {a | b:>3}  (binary: {bin(a | b)})")
print(f"a ^ b  (XOR):  {a ^ b:>3}  (binary: {bin(a ^ b)})")
print(f"~a     (NOT):  {~a:>3}  (binary: {bin(~a)})")
print(f"a << 2 (LSHIFT): {a << 2:>3}  (binary: {bin(a << 2)})")
print(f"a >> 1 (RSHIFT): {a >> 1:>3}  (binary: {bin(a >> 1)})")

# Practical: check if a number is even using bitwise AND
num = 42
is_even = (num & 1) == 0
print(f"\n{num} is even: {is_even} (using bitwise AND with 1)")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Slide 23: Best Practices (PEP 8)</span>

<span style="font-family: 'Times New Roman', Times, serif;">

PEP 8 is Python's official style guide. Key recommendations:

1. **Use descriptive variable names** -- `student_name` not `sn`.
2. **Use snake_case** for variables and functions: `my_variable`, `calculate_total()`.
3. **Use UPPER_CASE** for constants: `MAX_SIZE = 100`.
4. **Use 4 spaces** for indentation (never tabs).
5. **Limit lines to 79 characters**.
6. **Add spaces around operators**: `x = 1 + 2`, not `x=1+2`.

</span>

````python
# PEP 8 Best Practices -- Good vs. Bad examples

# BAD: unclear names, poor formatting
# x=5
# sn="Alice"
# def f(a,b): return a+b

# GOOD: descriptive names, proper spacing
MAX_RETRIES = 3  # Constant in UPPER_CASE

student_name = "Alice"              # snake_case for variables
student_age = 22                     # Descriptive name
grade_point_average = 3.85           # Clear and readable


def calculate_average(scores):       # snake_case for functions
    """Calculate the average of a list of scores."""
    if not scores:
        return 0.0
    return sum(scores) / len(scores)


# Proper spacing around operators
total = 10 + 20
is_valid = student_age >= 18 and grade_point_average >= 3.0

print(f"Student: {student_name}")
print(f"Age: {student_age}")
print(f"GPA: {grade_point_average}")
print(f"Valid: {is_valid}")

test_scores = [85, 92, 78, 95, 88]
print(f"Average: {calculate_average(test_scores):.2f}")
````

### <span style="font-family: 'Times New Roman', Times, serif;">Assignment 2: Build a Simple Calculator</span>

<span style="font-family: 'Times New Roman', Times, serif;">

**Objective:** Create a calculator program that demonstrates your knowledge of arithmetic operators, comparison operators, input/output, and control flow.

**Requirements:**

1. **Prompt the user** for:
   - First number (convert to `float`)
   - An operator (`+`, `-`, `*`, `/`, `//`, `%`, `**`)
   - Second number (convert to `float`)

2. **Use `if-elif-else`** to perform the correct operation based on the operator.

3. **Handle division by zero** -- if the user tries to divide by zero, print an error message instead of crashing.

4. **Display results** using f-strings with appropriate formatting:
   - Show the full expression and result
   - For division, show the result to 4 decimal places
   - Show whether the result is positive, negative, or zero

5. **Implement all 7 arithmetic operators:** `+`, `-`, `*`, `/`, `//`, `%`, `**`

**Sample Output:**
```
===== SIMPLE CALCULATOR =====
Enter first number: 15
Enter operator (+, -, *, /, //, %, **): /

The result is positive.
==============================
```

</span>

````python
# ===== ASSIGNMENT 2: Simple Calculator =====
# TODO: Implement your solution below

# Step 1: Get user input
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, //, %, **): ")
num2 = float(input("Enter second number: "))

# Initialize a variable to hold the final answer
result = None

# Step 2 & 3: Use if-elif-else for operations and handle division by zero
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero.")
    else:
        result = num1 / num2
elif operator == "//":
    if num2 == 0:
        print("Error: Cannot floor divide by zero.")
    else:
        result = num1 // num2
elif operator == "%":
    if num2 == 0:
        print("Error: Cannot calculate modulo with zero.")
    else:
        result = num1 % num2
elif operator == "**":
    result = num1 ** num2
else:
    print(f"Error: Invalid operator '{operator}' entered.")

# Step 4: Display formatted results
if result is not None:
    print(f"\nResult: {num1} {operator} {num2} = {result:.2f}")

print("Complete Assignment 2 by implementing the simple calculator.")
````
