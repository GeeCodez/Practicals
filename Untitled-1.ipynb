### Conditional statements in python 
* Python uses indentations to define code blocks
* If statements are used for conditions and they result in booleans

```python
a = 3366
b = 200

if b > a:
  print("b is greater than a")
```

```python
a = 200
b = 33
if b > a:
    print("b is greater than a")
else:
    print("b is not greater than a")
```

### Ternary operators
value_if_true if condition else value_if_false

```python
name = input("Enter your name: ")
print("Hello, " + name ) if name else print("Who are you")
```

```python
x = 1
while x <= 5:
    print(x)
    x += 1
```
