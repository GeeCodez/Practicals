```python
# Student grade management system

students = [
    {"name": "Henry Mettle", "scores": [92, 88, 95, 78, 90]},
    {"name": "Ama Serwaa", "scores": [65, 72, 58, 80, 70]},
    {"name": "Kojo Owusu", "scores": [85, 87, 83, 90, 88]}
]

for s in students:
    s["average"] = sum(s["scores"]) / len(s["scores"])

    avg = s["average"]
    if avg >= 90:
        s["grade"] = "A"
    elif avg >= 80:
        s["grade"] = "B"
    elif avg >= 70:
        s["grade"] = "C"
    elif avg >= 60:
        s["grade"] = "D"
    else:
        s["grade"] = "F"

honour_roll = [s["name"] for s in students if s["average"] >= 85]

ranked = sorted(students, key=lambda s: s["average"], reverse=True)

print("===== STUDENT GRADE REPORT =====")
print("Name            Avg     Grade")
print("-" * 35)

for s in students:
    print(f"{s['name']:<15} {s['average']:<7.2f} {s['grade']}")

print("\nHonour Roll:", ", ".join(honour_roll))

print("\nRankings:")
for i, s in enumerate(ranked, 1):
    print(f"#{i} {s['name']} - {s['average']:.2f}")
```

```python
numbers_list = [4, 7, 2, 9, 4, 1, 7, 3, 8, 2, 5, 9]
second_numbers_list = [1, 3, 5, 7, 11, 13]

unique_numbers = set(numbers_list)

duplicate_numbers = {number for number in numbers_list if numbers_list.count(number) > 1}

second_unique_numbers = set(second_numbers_list)

common_numbers = unique_numbers & second_unique_numbers
only_first_list = unique_numbers - second_unique_numbers
only_second_list = second_unique_numbers - unique_numbers

even_numbers = [number for number in numbers_list if number & 1 == 0]
odd_numbers = [number for number in numbers_list if number & 1 == 1]

def calculate_statistics(*numbers):
    return {
        "sum": sum(numbers),
        "count": len(numbers),
        "average": f"{sum(numbers) / len(numbers):.2f}",
        "min": min(numbers),
        "max": max(numbers)
    }

statistics_result = calculate_statistics(*numbers_list)

frequency_dictionary = {number: numbers_list.count(number) for number in set(numbers_list)}

index_pairs = list(zip(range(len(numbers_list)), numbers_list))

print("===== NUMBER ANALYSIS =====")
print("Original:", numbers_list)
print("Unique:", unique_numbers)
print("Duplicates:", duplicate_numbers)
print("\nEven (bitwise):", even_numbers)
print("Odd (bitwise):", odd_numbers)
print("\nStats:", statistics_result)
print("\nFrequency:", frequency_dictionary)
print("\nIndex pairs:", index_pairs)
```
