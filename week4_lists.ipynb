```python
ITA=[1,2,3,4,5]
```

```python
# len is used to get the length of a list
len(ITA)
```

```python
# indexing is used to get a specific item from a list
ITA[3]
```

```python
# append is used to add an element to the end of a list
ITA.append(6)
len(ITA)
```

```python
print(ITA)
```

```python
ITA.remove(3)
```

```python
print(ITA)
```

```python
# slicing is used to get a range of items from a list
# takes elements from the start index specified to 1-before the end index specified
# if the start index is not specified, it defaults to 0
ITA[1:4]
```

* A list includes **Variable name**, **assignment operator** and **list items**.
* Indexing starts from 0
